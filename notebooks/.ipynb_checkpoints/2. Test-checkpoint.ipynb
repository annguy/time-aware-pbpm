```python
F = Datahandler()
name=F.read_data(eventlog)
model_name=input("Enter the model path e.g model/sub_path/model_name.h5: ")
print('  ')
#Reading the Data

if 'product' in F.spamread.columns:
    num_features=7
    print('Context Data')
    factor=F.context_factor()
    spamreader,max_task = F.log2np(hidden_vector)
    D=Preprocess_Context()
    divisor,divisor2,divisor3 = D.divisor_cal(spamreader)
    maxlen,chars,target_chars,char_indices,indices_char,target_char_indices,target_indices_char= D.dict_cal()
    lines,caseids,lines_t,lines_t2,lines_t3,contexts=D.test_set()
    test=input('Enter 1 for next_step eval,2 for remaining_step eval: ')

    if test =='1':        
        Nxt=NextStep_Context(lines,caseids,lines_t,lines_t2,lines_t3,contexts,maxlen,eventlog,chars,target_chars,divisor,divisor2,target_indices_char,char_indices,model_name,factor,num_features)
        print('Running Test')
        Nxt.test()
    elif test=='2':
        Rmn=Remaining_Step_Context(lines,caseids,lines_t,lines_t2,lines_t3,contexts,maxlen,eventlog,chars,target_chars,divisor,divisor2,divisor3,target_indices_char,char_indices,model_name,factor,num_features)
        print('Running Test')
        Rmn.test()



else:  
    print('No context')
    num_features=6
    spamreader,max_task = F.log2np(hidden_vector)
    D=Preprocess()
    divisor,divisor2,divisor3 = D.divisor_cal(spamreader)
    maxlen,chars,target_chars,char_indices,indices_char,target_char_indices,target_indices_char= D.dict_cal()
    lines,caseids,lines_t,lines_t2,lines_t3=D.test_set()
    test=input('Enter 1 for next_step eval,2 for remaining_step eval: ')

    if test =='1':        
        Nxt=NextStep(lines,caseids,lines_t,lines_t2,lines_t3,maxlen,eventlog,chars,target_chars,divisor,divisor2,target_indices_char,char_indices,model_name)
        print('Running Test')
        Nxt.test()
    elif test=='2':
        Rmn=Remaining_Step(lines,caseids,lines_t,lines_t2,lines_t3,maxlen,eventlog,chars,target_chars,divisor,divisor2,divisor3,target_indices_char,char_indices,model_name)
        print('Running Test')
        Rmn.test()
```
